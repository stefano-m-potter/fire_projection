```python
import json
import requests
import os
from base64 import b64encode

# --- 1. Configuration ---
username = "spotter@woodwellclimate.org"
password = "Stefano001!"
save_path = "/explore/nobackup/people/spotter5/clelland_fire_ml/new_firecci"
years = ["2021", "2022"]

# Ensure the target directory exists
if not os.path.exists(save_path):
    os.makedirs(save_path, exist_ok=True)

# --- 2. Get the Access Token ---
token_url = "https://services.ceda.ac.uk/api/token/create/"
auth_str = b64encode(f"{username}:{password}".encode('utf-8')).decode("ascii")
headers = {"Authorization": f'Basic {auth_str}'}

print("Requesting access token...")
response = requests.post(token_url, headers=headers)

if response.status_code == 200:
    access_token = response.json().get("access_token")
    print("Token acquired successfully.")
    
    # --- 3. Execute Wget for 2021 and 2022 ---
    base_url = "https://dap.ceda.ac.uk/neodc/esacci/fire/data/burned_area/MODIS/pixel/v5.1/compressed/"
    
    for year in years:
        print(f"\n--- Starting download for year: {year} ---")
        target_url = f"{base_url}{year}/"
        
        # wget flags:
        # -r: recursive
        # -np: no parent (don't go back up to other years)
        # -nd: no directories (keep local folder flat)
        # -P: specify the output directory
        # -c: continue (resumes if a download was interrupted)
        wget_cmd = (
            f'wget -e robots=off -r -np -nd -c '
            f'--header "Authorization: Bearer {access_token}" '
            f'-P {save_path} '
            f'{target_url}'
        )
        
        os.system(wget_cmd)
    
    print("\nAll downloads for 2021 and 2022 are complete.")
else:
    print(f"Failed to get token: {response.status_code}")
    print(response.text)
```
