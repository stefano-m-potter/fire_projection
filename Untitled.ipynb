```python
import os
import numpy as np
import rasterio as rio
from rasterio.mask import mask
import geopandas as gpd
import matplotlib.pyplot as plt
from tqdm import tqdm

# ---------------- PATHS ----------------
FIRECCI_ANNUAL_DIR = "/explore/nobackup/people/spotter5/clelland_fire_ml/fire_cci_us"
ECOS_PATH = "/explore/nobackup/people/spotter5/helene/raw/merge_eco_v2.shp"

# -------------- PARAMETERS --------------
YEARS = list(range(2001, 2023))

def main():
    print(f"Loading ecoregions shapefile: {ECOS_PATH}")
    # Load shapefile
    gdf = gpd.read_file(ECOS_PATH)
    
    years_processed = []
    burned_area_mha = []
    
    # Track the CRS of the first raster to avoid reprojecting the shapefile every loop
    target_crs = None
    geometries = None

    for year in tqdm(YEARS, desc="Processing Years"):
        tif_path = os.path.join(FIRECCI_ANNUAL_DIR, f"{year}.tif")
        
        if not os.path.exists(tif_path):
            print(f"\n[WARN] Missing file for {year}: {tif_path}")
            continue
            
        with rio.open(tif_path) as src:
            # 1. Match CRS and extract geometries (only needs to run once)
            if target_crs is None:
                target_crs = src.crs
                if gdf.crs != target_crs:
                    print(f"Reprojecting shapefile to match raster CRS: {target_crs}")
                    gdf = gdf.to_crs(target_crs)
                
                # Extract geometries to pass to the masking function
                geometries = gdf.geometry.values
            
            # 2. Clip the raster to the shapefile geometry
            try:
                # crop=True subsets the raster extent to the bounding box of the shapefile
                out_image, out_transform = mask(src, geometries, crop=True)
            except ValueError:
                # Triggers if the geometries do not overlap the raster at all
                print(f"\n[SKIP] No overlap found for {year}")
                continue
            
            # 3. Process the clipped array
            arr = out_image[0] # Band 1
            
            # Count burned pixels (any value > 0)
            burned_pixel_count = np.count_nonzero(arr > 0)
            
            # Calculate single pixel area (res[0] is X res, res[1] is Y res)
            pixel_area_sq_meters = abs(src.res[0] * src.res[1])
            
            # 4. Convert area to Millions of Hectares (Mha)
            # 1 Hectare = 10,000 sq meters. 1 Million Hectares = 10^10 sq meters.
            total_area_sq_meters = burned_pixel_count * pixel_area_sq_meters
            area_mha = total_area_sq_meters / 10_000_000_000
            
            years_processed.append(year)
            burned_area_mha.append(area_mha)

    # -------------- PLOTTING --------------
    if not years_processed:
        print("No data was processed. Exiting without plotting.")
        return

    plt.figure(figsize=(12, 6))
    plt.plot(years_processed, burned_area_mha, marker='o', linestyle='-', color='firebrick', linewidth=2)
    
    plt.title('Annual Burned Area (2001-2022)\nClipped to Ecoregions', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Burned Area (Millions of Hectares)', fontsize=12)
    
    # Make the x-axis display strictly the years we processed
    plt.xticks(years_processed, rotation=45)
    plt.grid(True, linestyle='--', alpha=0.6)
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------- PATHS ----------------
# Path to the CSV generated by your previous script
CSV_PATH = "/explore/nobackup/people/spotter5/clelland_fire_ml/burned_area_summaries/burned_area_by_ecoregion_predictions.csv"

def main():
    # 1. Load the CSV data
    try:
        df = pd.read_csv(CSV_PATH)
        print(f"Successfully loaded data from: {CSV_PATH}")
    except FileNotFoundError:
        print(f"[ERROR] File not found: {CSV_PATH}")
        return

    # 2. Verify required columns exist
    required_cols = ['year', 'ba_mcd_native_Mha', 'ba_firecci_native_Mha']
    missing_cols = [c for c in required_cols if c not in df.columns]
    if missing_cols:
        print(f"[ERROR] Missing required columns in CSV: {missing_cols}")
        return

    # 3. Aggregate data: Group by 'year' and sum the areas across all ecoregions
    annual_totals = df.groupby('year')[['ba_mcd_native_Mha', 'ba_firecci_native_Mha']].sum().reset_index()

    # 4. Plot the results
    plt.figure(figsize=(10, 6))

    # Plot MCD64A1
    plt.plot(
        annual_totals['year'], 
        annual_totals['ba_mcd_native_Mha'], 
        marker='o', 
        linestyle='-', 
        linewidth=2,
        label='MCD64A1',
        color='tab:blue'
    )

    # Plot FireCCI
    plt.plot(
        annual_totals['year'], 
        annual_totals['ba_firecci_native_Mha'], 
        marker='s', 
        linestyle='--', 
        linewidth=2,
        label='FireCCI',
        color='tab:orange'
    )

    # 5. Format the plot for readability
    plt.title('Total Annual Burned Area Across All Ecoregions (2001-2022)', fontsize=14, fontweight='bold')
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Total Burned Area (Millions of Hectares)', fontsize=12)
    
    # Ensure every year is displayed on the x-axis
    plt.xticks(annual_totals['year'], rotation=45)
    
    # Add grid and legend
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.legend(fontsize=11)
    
    # Adjust layout so labels don't get cut off
    plt.tight_layout()

    # Display the plot directly (no saving)
    plt.show()

if __name__ == "__main__":
    main()
```
